## Summary
This code creates the indicator we will use as signal for the bt strategy. The indicator is created as the difference between positive and negative sentiment (Positive-Negative). Therefore, if the indicator is greater than 0, it means during that day the overall sentiment is positive. On the other hand, if the indicator is lower than 0, it means during that day the overall sentiment was negative

```python
import bt 
import pandas as pd
import numpy as np


def ticker_def(tickers, start, end):
    data_bt = bt.get(tickers, start=start, end=end)
    return data_bt

def indicator_sent(df_sentana):
    indicator = df_sentana.pivot(index='Date',columns='sentiment', values='polarity_norm')
    indicator[indicator.isnull()]=0
    indicator['signal']=indicator['positive']-indicator['negative']
    indicator = indicator.reset_index()
    indicator['Date']= indicator["Date"].astype(np.datetime64)
    indicator = indicator.set_index('Date')
    indicator.sort_index(inplace=True)
    rng = ticker_def('hgen', '01/23/2017', end = '2021-06-07')
    indicator_df = pd.concat([rng, indicator], axis=1)
    indicator_df.dropna(subset=['hgen'],inplace=True)
    return indicator_df

df_sent = pd.read_csv('.\sentimentanalysis_count$$HGEN OR Humanigen.csv')
indicator = indicator_sent(df_sent)

indicator.to_csv('.\signal_hgen_2.csv',encoding='utf-8-sig') 
```
